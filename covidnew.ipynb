```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd 
import seaborn as sns
from datetime import date
%matplotlib inline
```

```python
sns.set_context(context="notebook")
```

```python
covid = pd.read_csv("covid_19_clean_complete.csv")
```

```python
covid.info()
```

```python
covid.head()
```

```python
covid["Month"] = covid["Date"].apply(lambda x : x.split("/")[0])
covid["Day of Month"] = covid["Date"].apply(lambda x : x.split("/")[0])
```

```python
covid.head()
```

```python
covid["Month"].max()
```

```python
covid["Month"].min()
```

```python
def days_passed(datestring):
    d=datestring.split("/")
    #year,month,date
    #set d0 as starting date in dataset
    d0 = date(2020,1,22)

    d1 = date(2020,int(d[0]),int(d[1]))
    delta = d1-d0
    return delta.days
```

```python
covid["Days Passed"] = covid["Date"].apply(days_passed)
```

```python
covid["Days Passed"].max()
```

```python
sns.lineplot(x="Days Passed",y="Confirmed",data=covid,size=20)
```

```python
covind = covid[(covid["Country/Region"]=="India") & (covid["Confirmed"]>0)]
```

```python
covind["Days Passed"] = covind["Days Passed"] - covind["Days Passed"].iloc[0]
```

```python
covind
```

```python
sns.lineplot(x="Days Passed",y="Confirmed",data=covind)
```

```python
groups = covid.groupby("Country/Region")
```

```python
country_df = groups.describe()
```

```python
country_df["Fatality Rate"] = country_df["Deaths"]["max"] / country_df["Confirmed"]["max"]
```

```python
x = country_df[["Confirmed","Fatality Rate","Deaths"]]
```

```python
x
```

```python
x.reset_index(inplace=True)
```

```python
sns.barplot(x="Country/Region",y="Fatality Rate",data=x[x["Confirmed"]["max"]>10000],errwidth=10)
```
